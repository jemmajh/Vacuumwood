```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/jemmajeonghan/Documents/FinalProject/financialmodel/VerticalFarmModel/data/electricity_prices_2013_2020.csv")
df["timestamp"] = pd.to_datetime(df["timestamp"])
df["hour"] = df["timestamp"].dt.hour
df["month"] = df["timestamp"].dt.month
```

```python
df.groupby("hour")["price_eur_per_kwh"].mean().plot()
plt.title("Average Price by Hour")
plt.show()
```

```python
df.groupby("month")["price_eur_per_kwh"].mean().plot()
plt.title("Average Price by Month")
plt.show()
```
